```python
report_date = crime_df["REPORTED_DATE"].apply(lambda x: x.split(" "))
crime_df["report_m/d/y"] = FIRST_OCCURRENCE_DATE.apply(lambda x: x[0])
crime_df["report_time"] = FIRST_OCCURRENCE_DATE.apply(lambda x: x[1])
crime_df["report_am/pm"] = FIRST_OCCURRENCE_DATE.apply(lambda x: x[2])
crime_df["report_12-time"]= crime_df[["first_occur_time","first_occur_am/pm"]].apply(lambda x: "".join(x),axis = 1)
crime_df
```

```python
# the date format is not xx-xx-xxxx 5/6/2017 and can not run.... 
from datetime import datetime

def days_between(row):
    d1 = datetime.strptime(row["first_occur_m/d/y"], "%m/%d/%y")
    d2 = datetime.strptime(row["report_m/d/y"], "%m/%d/%y")
    return abs((d2 - d1).days)

crime_df["days_report"] = crime_df.apply(days_between,axis =1 )

crime_df
```
