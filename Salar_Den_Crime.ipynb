```python
import csv
import pandas as pd
```

```python
crime_csv = 'denver-crime-data/crime.csv'
```

```python
crime_df = pd.read_csv(crime_csv)
crime_df.head()
```

```python
TYPE_ID = crime_df.groupby(['OFFENSE_TYPE_ID']).count()

TYPE_ID.head(10)
```

```python
TYPE_ID.sort_values(by = ['INCIDENT_ID'], axis=0, ascending=False, inplace=False)
TYPE_ID.head()
```

```python
neighborhood_id = crime_df.groupby(['NEIGHBORHOOD_ID']).count()
neighborhood_id.sort_values(by = ['INCIDENT_ID'], axis=0, ascending=False, inplace=False)
neighborhood_id.head(10)
```
